```python
df =df.fillna(0)
df
```

```python
df.iloc[10]
df.rename(columns={'pr1':'P1'}, inplace = True)
df.rename(columns={'pr2':'P2'}, inplace = True)
df.rename(columns={'pr3':'P3'}, inplace = True)
df.head()
```

```python
df['Average'] = (df['P1']+df['P2']+df['P3'])/3
df['Average'] = df['Average'].round(2)
df.head()
```

```python

print("To print data in sorted order")
df = df.sort_values(by="Total", ascending = False)
df = df.sort_values('name')
df
```

```python
female_count = (df['gender'] == 'Female').sum()
print("Total no. of Male candidates:",female_count)
male_count = (df['gender'] == 'Male').sum()
print("Total no. of Male candidates:",male_count)
urban_total = (df['area'] == 'Urban').sum()
print("Total Urban Areas:", urban_total)
rural_Total = (df['area'] == 'Rural').sum()
print("Total Urban Areas:", rural_Total)
```

```python
print("Average Total score by Gender")
print(df.groupby('gender')['Total'].mean())
print("\nAverage Total score by Area")
print(df.groupby('area')['Total'].mean())
print("\nTotal Students by area :")
for area,total in df.groupby('area')['Total'].sum().items():
    print(f"{area}:{total}")

print("\nTotal students by gender: ")
for gender,total in df.groupby('gender')['Total'].sum().items():
    print(f"{gender}:{total}")

print("\nAverage students by gender: ")
for gender, Avg in df.groupby('gender')['Average'].mean().items():
    print(f"{gender}:{Avg}")
    
print("\nAverage students by Area: ")
for area, Avg in df.groupby('area')['Average'].mean().items():
    print(f"{area}:{Avg}")

print("Total score of students by area")
print(df.groupby('area')[['P1','P2','P3','Total']].mean())
```

```python
print("counts total Rural and Urban:",df['area'].value_counts())
print("counts Area: ",df['area'].count())
```

```python
print("Person whose gender is Male and Total > 200")
df.query('gender == "Male"  and Total >200')
```

```python
#define a grade calculator
def grade(x):
    if x>450: return 'A'
    elif x>350: return 'B'
    elif x>250: return 'C'
    else: return 'D'

df['Grade'] = df['Average'].apply(grade)
df
```

```python
print("create a pivot table" )
pivot = pd.pivot_table(
    df,
    values='Total',
    index='area',
    columns='gender',
    aggfunc='sum',
)
print(pivot)

print("\n Multi- level Grouping")
grouped = df.groupby(['area','gender'])['Total'].agg(['mean','count','max','min'])
print(grouped)
```

```python
#Sorting with multiple columns
df.sort_values(['area','Total'], ascending = [True, False], inplace = True)
print(df)
```

```python
df.rename(columns= {'name':'Name','gender':'Gender','area':'Area'}, inplace = True)
df
```

```python
#Subplots (multiple charts together)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
df['Total'].plot(kind='hist', bins=10, color='orange', ax=axes[0], title='Total Distribution')
df['Area'].value_counts().plot(kind='bar', color='teal', ax=axes[1], title='Count by Area')
plt.tight_layout()
```

```python
#3️⃣ Line Plot for Trends
# If you had test scores as time series per student:
df[['P1', 'P2', 'P3']].mean().plot(kind='line', marker='o', color='red')
plt.title("Average Scores Across Tests")
plt.xlabel("Test")
plt.ylabel("Average Score")
plt.grid(True)
plt.show()
```

```python
avg_gender = df.groupby("Gender")["Total"].mean()
plt.bar(avg_gender.index, avg_gender.values, color=["blue","pink"])
plt.xlabel("Gender")
plt.ylabel("Average Total Score")
plt.title("Gender-wise Performance")
plt.show()
```

```python
#Line chart
avg_score = df.groupby("Name")[["P1","P2","P2"]].mean().plot(kind="line", marker="d")
plt.title("Urban vs Rural Average Scores")
plt.xlabel("Performance Records")
plt.ylabel("Average Marks")
plt.grid(True)
plt.legend(title="Area")
plt.show()

```

```python
#Scatter plot
plt.scatter(df["P1"], df["P2"], c="purple", alpha=0.7)
plt.xlabel("P1 Marks")
plt.ylabel("P2 Marks")
plt.title("Correlation between PR1 and PR2")
plt.grid(True)
plt.show()
```
