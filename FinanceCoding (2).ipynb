```python
import pandas as pd
```

```python
df=pd.read_excel("Finance_sheet.xlsx")
df
```

```python
print("Top 5 Records")
print(df.head())                # First 5 rows
print("Bottom 5 Records")
print(df.tail())                # Last 5 rows
print("Shape")
print(df.shape)                 # (rows, columns)
print("Column Name")
print(df.columns)               # Column names
print("Data Type and Non Null Values")
print(df.info())                # Data types and non-null values
print("Data Types")
print(df.dtypes)                # Data types per column
```

```python
#Data Cleaning
df.isnull().sum()   
```

```python
df.dropna() 
```

```python
df.fillna(0)    
```

```python
df['CONTACT'] = df['CONTACT'].astype(str)
df
```

```python
df.columns = df.columns.str.strip() # Strip column names of whitespace
df
```

```python
#Accessing Data
df['NAME']               # Access column
df[['NAME', 'GENDER']]   # Access multiple columns
df.iloc[0]               # First row
df.loc[0, 'NAME']        # Value from row 0, column 'NAME'
df[df['GENDER'] == 'Male']   # Filter condition
```

```python
#Grouping and Aggregation
df.groupby('GENDER').size()                             # Count per gender
df.groupby('BANK')['AMOUNT'].sum()                      # Total amount per bank
df.groupby(['BANK', 'GENDER'])['AMOUNT'].mean()         # Avg amount by bank and gender
```

```python
#Sorting and Filtering
df.sort_values('AMOUNT', ascending=False)               # Sort by amount
df[df['AMOUNT'] > 100000]                               # Filter high loans
df[(df['GENDER'] == 'Female') & (df['AMOUNT'] > 50000)] # Multiple filters
```

```python
#Creating New Columns
df['Interest'] = (df['AMOUNT'] * df['Rate'] * df['TIME']) / 100  # Simple interest
df['Full_Name'] = df['NAME'].str.upper() 
df
```

```python
#Value Counts and Unique Values
df['BANK'].value_counts()       # Count of records per bank
df['GENDER'].unique()           # Unique values in GENDER
df['GENDER'].nunique()          # Number of unique genders
```

```python
#Pivot Tables
pd.pivot_table(df, values='AMOUNT', index='BANK', columns='GENDER', aggfunc='sum')
```

```python
#Correlation and Stats
df.corr(numeric_only=True)     # Correlation between numeric fields
df['AMOUNT'].mean()            # Average loan amount
df['AMOUNT'].max()             # Max loan amount
```

```python
#Handling Duplicates
df.duplicated().sum()          # Count of duplicates
df.drop_duplicates()           # Remove duplicates
```

```python
#Common Analytics Use-Cases
#1.Total loan disbursed
df['AMOUNT'].sum()
```

```python
#Number of loans per bank	
df['BANK'].value_counts()
```

```python
#Gender distribution
df['GENDER'].value_counts()
```

```python
#Average EMI by bank
df.groupby('BANK')['EMI'].mean()
```

```python
#ROUND THE VALUES
df['EMI'] = df['EMI'].round(2)
df
```

```python
#Round of Multiple Columns
df = df.round({
    'Rate': 2,
    'AMOUNT': 0,
    'EMI': 0
})
df
```

```python
#Round While Creating a New Column
#For example, if you're calculating interest:
df['Interest'] = ((df['AMOUNT'] * df['Rate'] * df['TIME']) / 100).round(2)
df
```

```python
#Use of Logical Operator in Pandas
df[(df['GENDER'] == 'Female') & (df['BANK'] == 'HDFC')]
```

```python
#Filter: Customers whose name starts with "A"
df[df['NAME'].str.startswith('A')]
```

```python
#Storing the Filtered Data
filtered_df = df[(df['BANK'] == 'SBI') & (df['Rate'] < 10)]
filtered_df.to_excel('sbi_low_rate_loans.xlsx', index=False)
```

```python
#Negation Example (NOT)
#Get all records except those with time period = 5 years:
df[~(df['TIME'] == 5)]
```

```python
#Not from ICICI bank
df[df['BANK'] != 'ICICI']
```

```python
#Filter: Loan between ₹50,000 and ₹1,00,000
df[(df['AMOUNT'] >= 50000) & (df['AMOUNT'] <= 100000)]
```

```python
#Filter: Missing Aadhar numbers
df[df['AADHAR'].isnull()]
```

```python
#Replace Values
df['BANK'] = df['BANK'].replace({'SBI': 'State Bank of India', 'ICICI': 'ICICI Bank'})
df
```

```python
#Drop Columns or Rows
df.drop(columns=['AADHAR'], inplace=True)   # Drop a column
df.drop(index=[0, 1], inplace=True)    
```

```python
df
```

```python
#Filter Using isin()
df[df['BANK'].isin(['SBI', 'HDFC'])]   
```

```python
#Conditional Column with np.where() (alternative to apply)
import numpy as np
df['EMI_Status'] = np.where(df['EMI'] > 2000, 'High EMI', 'Low EMI')
df
```

```python
#Query Syntax (Readable Filter)
df.query("GENDER == 'Female' and BANK == 'PNB'")
```
