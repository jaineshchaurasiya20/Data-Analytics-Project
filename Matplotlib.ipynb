```python
import matplotlib.pyplot as plt
import pandas as pd
x = [1,2,3,4,5]
y = [20, 10, 30, 20, 20]
# Line plot
explode=[0,0,0.1,0,0]
plt.title("Line Chart")
plt.plot(x,y,color='blue', marker='D')
plt.show()

# Pie chart
plt.pie(y,labels=x, autopct='%2.1f%%',explode=explode)
plt.show()

#Line chart
plt.title("MUHAHAHAHAHAHAHAH.....")
plt.plot(x,y,color='red',linestyle=':')
plt.show()

#Bar chart
a = ['A', 'B','C', 'D']
b = [10, 20, 30, 50]
plt.bar(a,b, color ='green')
plt.title("Bar chart")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()

# Scatter chart
_a1 = [7,5,4,3,2,2,1]
_b1 = [56,34,42,50,7,8,2]
plt.scatter(_a1,_b1, marker = '*', s=500,color = "blue", edgecolors='red')
plt.show()


#city temperature
city = ['Lko','Delhi', 'Mumbaia', 'kolkata']
temp = [ 45.5, 34.4, 67.34, 12.9]
plt.bar(city, temp)
plt.title("City- Temperature graph")
plt.xlabel("City")
plt.ylabel("Temperature")
plt.show()

# Multiliple line on same graph
x = [1,2,3,4,5]
y1 =[2,4,6,8,10]
y2 = [1,3,5,7,9]
plt.plot(x,y1,label = 'Line 1', color = 'blue')
plt.plot(x,y2, label = 'Line 2', color = "Red")
plt.title("Multiple Line")
plt.legend() #show color label
plt.show()

df = pd.DataFrame({'x': [1, 2, 3,4,5], 'y': [4, 5, 6,50, 40]})
df.plot(kind='line', x='x', y='y', color='skyblue', marker='o', )  # 'edgecolors' is equivalent to 'mec'
plt.show()
```

```python
# Matplotlib with numpy
import numpy as np
x = np.array([0,6])
y = np.array([0,100])
plt.plot(x,y)
plt.title("Mat with numpy")
plt.show()

a = np.array([1,2,3,4,5])
b = np.array([5,6,3,6,6])
plt.plot(a,b)
plt.show()

points = np.array([3,7.5,9,10,20,10,5])
plt.plot(points, marker = 'D', ms = 25, mfc ='r', mec = 'g')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = { 'Student': ['Jai', 'Jainesh', 'Jack', 'Jacky', 'Janu'],
         'English': [70, 60, 90, 100, 99],
         'Maths': [100, 99, 98, 97, 95],
         'Science': [80, 89, 98, 76, 85]}
df = pd.DataFrame(data)
print(df)
df['Total'] = df[['English', 'Maths', 'Science']].sum(axis = 1)
print(df)

plt.bar(df['Student'], df['Total'])
plt.show()
```
