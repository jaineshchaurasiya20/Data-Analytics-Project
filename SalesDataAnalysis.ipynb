```python
df = pd.read_csv("Sales_data (1).csv")
df.head()
```

```python
#Sales vs Profit
sb.lineplot(x='profit_margin', y='sales',data=df)
plt.title("Line plot between Sales vs Profit")
plt.show()
```

```python
#Monthly sales trend over time
df['Order Date'] = pd.to_datetime(df['order_date'])
df['YearMonth'] = df['Order Date'].dt.to_period('M').astype(str)

monthly = df.groupby(['YearMonth', 'region'])['sales'].sum().reset_index()

sb.relplot(data=monthly, x='YearMonth', y='sales', kind='line', col='region', col_wrap=2, marker='o', height=4, aspect=1.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Monthly sales trend over time
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.resample('ME', on='Order Date')['sales'].sum().reset_index()

sb.lineplot(data=monthly_sales, x='Order Date', y='sales', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Profit trends across quarters
df['Quarter'] = df['Order Date'].dt.to_period('Q').astype(str)
quarterly_Profit = df.groupby('Quarter')['profit'].sum().reset_index()

sb.lineplot(data=quarterly_Profit,x='profit', y='Quarter',  marker='o')
plt.title('Profit trends across quarters')
plt.show()
```

```python
#Sales Trend by Region Over Time

monthly_region_sales = df.groupby(['YearMonth', 'region'])['sales'].sum().reset_index()

sb.lineplot(data=monthly_region_sales, x='YearMonth', y='sales', hue='region', marker='o')
plt.title('Sales Trend by Region Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Quantity sold per month
monthly_Quantity = df.groupby('YearMonth')['quantity'].sum().reset_index()
sb.barplot(data=monthly_Quantity, x='YearMonth', y='quantity')
plt.title("Quantity sold per month")
plt.xticks(rotation=80)
plt.show()
```

```python
#Discount trend over time
discount_trend = df.groupby('YearMonth')['discount'].sum().reset_index()
sb.lineplot(data=discount_trend, x='YearMonth', y='discount', marker='+')
plt.title("Discount trend over time")
plt.xticks(rotation=80)
plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Sample data with lat/lon
regions = ['East', 'West', 'Central', 'South']
sales = [12000, 9500, 8700, 11000]
lats = [26.8, 19.1, 23.0, 15.3]
lons = [82.6, 72.9, 78.5, 77.6]

# Create map
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([68, 98, 6, 38], crs=ccrs.PlateCarree())  # India bounding box

# Add features
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.LAKES)
ax.add_feature(cfeature.RIVERS)

# Plot sales points
for i in range(len(regions)):
    ax.plot(lons[i], lats[i], marker='o', color='red', markersize=sales[i]/1000, transform=ccrs.PlateCarree())
    ax.text(lons[i]+0.5, lats[i]+0.5, regions[i], transform=ccrs.PlateCarree())

plt.title('Sales by Region in India')
plt.show()
```
