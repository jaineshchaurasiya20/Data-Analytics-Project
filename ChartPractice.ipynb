```python
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sb
```

```python
#set theme
sb.set_style("whitegrid")#darkgrid/white,dark,ticks
df = sb.load_dataset("tips")
print(df.head())
```

```python
#distribution plot(histogram plot)
sb.histplot(df['total_bill'], bins=20, kde=True)
plt.show()
```

```python
#Box plot(Outlier Detection)
sb.boxplot(x="day",y="total_bill", data=df, hue="sex")
plt.title("Box plot for total bill")
plt.show()
```

```python
sb.kdeplot(data=df['total_bill'], fill=True, color='green')
plt.title("KDE Plot")
plt.show()
```

```python
#barplot
sb.barplot(x="day", data=df, hue="sex", estimator=sum)
plt.title("Bar plot")
plt.show()
```

```python
sb.countplot(x="day", data=df, hue="sex")
plt.show()
```

```python
sb.scatterplot(x="total_bill", y="size",data=df, hue="day", size="tip")
plt.show()
```

```python
sb.pairplot(df, hue="day")
plt.show()
```

```python
#heatmap(Correlation)
c = df.corr(numeric_only=True)
sb.heatmap(c, annot=True, cmap="coolwarm")
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')
sb.heatmap(data=numeric_df.corr(), cmap='coolwarm')
plt.show()
```

```python
sb.violinplot(x="day", y="total_bill", data=df, hue="smoker")
plt.show()
```
